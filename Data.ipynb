```python
import os
import pandas as pd

def update_csv_with_recommendations(csv_path, recommendation_folder, output_csv_path):
    # Read the NMF.csv file
    df = pd.read_csv(csv_path)

    # Get a list of track IDs from the MP3 files in the Recommendation folder
    recommendation_ids = []
    for filename in os.listdir(recommendation_folder):
        if filename.endswith(".mp3"):
            track_id = int(filename[:-4].lstrip('0'))  # Assuming the ID is the filename without the extension
            recommendation_ids.append(track_id)

    # Update the NMF.csv DataFrame to include only entries with IDs in the Recommendation folder
    updated_df = df[df['track_id'].isin(recommendation_ids)]

    # Save the updated DataFrame to the final_cut.csv file
    updated_df.to_csv(output_csv_path, index=False)

if __name__ == "__main__":
    csv_path = r"C:\Users\yolos\Downloads\GradProject2 - Copy (2)\RecommendationSystem\NMF.csv"  # Replace with the actual path to your NMF.csv file
    recommendation_folder = r"C:\Users\yolos\OneDrive\Desktop\Recommendation"  # Replace with the actual path to your Recommendation folder
    output_csv_path = "final_cut.csv"  # Replace with the desired path for the new CSV file

    update_csv_with_recommendations(csv_path, recommendation_folder, output_csv_path)
```
